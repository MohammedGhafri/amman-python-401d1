# Seaborn vs Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
def plot_lines(data1, data2):
    plt.figure()
    plt.plot(data1, label="Mocks x1")
    plt.plot(data2, label="Mocks x2")
    plt.title('Some Mock Data')
    plt.xlabel('Indeces')
    plt.ylabel('The mocks')
    plt.legend()
    plt.show()
plot_lines([3,7,14,9,4,1], [1, 13])
```

### Getting flights data from Seaborn

```python
flights = sns.load_dataset('flights')
```

```python
flights.head(5)
```

```python
sns.relplot(x='year', y='passengers', data=flights);
```

```python
year_sums = flights.groupby('year', as_index=False).sum()
year_mean = flights.groupby('year', as_index=False).mean()
```

```python
year_mean
```

```python
sns.relplot(x='year', y='passengers', data=year_sums);
```

```python
sns.relplot(x='year', y='passengers', data=year_mean);
```

```python
sns.lmplot(x="year", y="passengers", data=year_sums);
```

```python
graph = sns.barplot(x="month", y="passengers", data=flights)
```

```python
graph = sns.relplot(x="month", y="passengers", data=flights)
```

```python
graph = sns.barplot(x="year", y="passengers", data=flights)
```

```python
graph = sns.barplot(x="month", y="passengers", data=flights, ci=None)
```

```python
from calendar import month_abbr

new_graph = sns.barplot(x="month", y="passengers", data=flights, ci=None)

new_graph.set_xticklabels(labels=list(month_abbr)[1:]);
```
