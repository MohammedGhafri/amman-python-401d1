# Linear Regression Demo

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
salary = pd.read_csv('./salary.csv')
salary.head()
```

### Select features (alla viariables) for the prediction model

```python
X = salary['YearsExperience'].values
y = salary['Salary'].values
```

```python
X = X.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2 ,random_state=100)
```

```python
x_train.shape # 80% of X data
```

```python
x_test.shape # 20% of X data
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Yearly Salary')
plt.title('Training Data')
plt.show()
```

### Let's create the model and train it

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

#### Now testing:

```python
y_predict = model.predict(x_test)
```

```python
y_test # The leftover testing data for y
```

```python
y_predict # The resulted predicted data of y
```

#### Given a new years of experience, what is the predicted salary?

```python
new_person_years_exp = np.array([3]).reshape(-1,1)
new_person_years_exp
```

```python
model.predict(new_person_years_exp) # Predicted Salary for 3 years of experience
```

```python
new_person_years_exp = np.array([7]).reshape(-1,1) # Another person with 7 years of experience
```

```python
model.predict(new_person_years_exp) # Predicted Salary for 7 years of experience
```

### Accuracy of model

```python
model.score(x_train, y_train)*100
```

```python
model.score(x_test, y_test)*100
```

```python
### Plotting the predicted output vs. the test output
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_test, y_predict, color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Yearly Salary')
plt.title('Training Data vs Model')
plt.show()
```

### Repeat the process for Snacks feature

```python
salary = pd.read_csv('./salary_plus_snacks.csv')
salary.head()
X = salary['Snacks'].values
y = salary['Salary'].values
X = X.reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2 ,random_state=100)
```

```python
y_predict
```

```python
y_test
```

```python
model.score(x_train, y_train)*100 # Accuracy of trained data
```

```python
model.score(x_test, y_test)*100 # Accuracy of tested data
```
